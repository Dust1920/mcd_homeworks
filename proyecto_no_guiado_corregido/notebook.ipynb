## Pregunta 1:
En este caso, debemos calcular el porcentaje de licencias de star wars con licencia. Primero descargamos las líbrerias requeridas.

```python
import pandas as pd
lego_sets = pd.read_csv('datasets/lego_sets.csv')
lego_sets.isna().sum()
```

Observamos que tenemos valores faltantes en set_num, la que define el código de un set válido. Por lo tanto quitaremos los valores faltantes.

```python
n_lego = lego_sets.shape[0]
na_max = n_lego * 0.05
na_max
```

Como el número de faltantes, excepto en 'num_parts' es menor al máximo de na que podemos eliminar, procedemos 

```python
lego_sets_clean = lego_sets.dropna(subset=['set_num', 'name', 'theme_name'])
lego_sets_clean.isna().sum()
```

Ahora, extraemos los id con licencia.

```python
parent_themes = pd.read_csv('datasets/parent_themes.csv')
licensed_themes = parent_themes[parent_themes['is_licensed']]['name']
licensed_themes
```

Ahora, filtraremos la base de datos de lego, con respecto a la existencia de la licencia.

```python
licensed = lego_sets_clean['parent_theme'].isin(licensed_themes)
licensed_sets = lego_sets_clean[licensed]
licensed_sets.head()
```

Finalmente, calculamos el porcentaje. 

```python
all_sets = len(licensed_sets)
star_wars_sets = len(licensed_sets[licensed_sets['parent_theme'] == 'Star Wars'])
ratio = star_wars_sets/all_sets
the_force = int(ratio*100)
print(the_force)
```

## Pregunta 2

Ahora, debemos encontrar el año en el que Star Wars NO fue el más popular. Para ello usaremos un pivot_table para poder contar cuantos sets hubo por licencia con respecto al año.

```python
licensed_pivot = licensed_sets.pivot_table(index='year', columns='parent_theme', values='set_num', aggfunc='count')
licensed_pivot.head()
```

Entonces, solo queda hacer el conteo y buscar en que año Star Wars no tuvo el mayor número de ventas.

```python
new_era = int(licensed_pivot[licensed_pivot['Star Wars'] < licensed_pivot.max(axis='columns')].index[0])
print(new_era)
```
